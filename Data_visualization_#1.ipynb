```python
# Exercize: Plot a funkcion f(x)= 17*x^3-25*x+28*x-100
```

```python
# Creating array for x
```

```python
# Def xmin, xmax, step
```

```python
xmin = -10
xmax = 10
step_size = 0.2
```

```python
number_of_steps = int((xmax - xmin)/step_size)
x = [xmin]
x_curent= xmin
```

```python
# Creating for cikle
```

```python
for step in range(number_of_steps):
    x_curent = round(x_curent + step_size, 1)
    x.append(x_curent)
print(x)
```

```python
# def funkcion
```

```python
def f(x):
    return 17 * x**3 - 25 * x + 28 * x - 100
```

```python
# f(x) calculation
```

```python
y = []
```

```python
for item in x:
    fx = round(f(item))
    #print(f'x =  {item}, f(x) =  {fx}')
    y.append(fx)
```

```python
# Plot data
```

```python
plt.figure(figsize=(20, 5), facecolor='#BBC6C8')
ax = plt.axes().set_facecolor('#E5E3E4')
plt.axvline(x=0, color='#5BA199', lw=1)
plt.axhline(y=0, color='#5BA199', lw=1)
plt.plot(x,y, color='#469597', lw=1, linestyle='-', marker='.')
plt.grid()
plt.title('Grafikas f(x)= 17*x^3-25*x+28*x-100', pad=20, fontweight='bold')
plt.xlabel('X values', fontweight='bold')
plt.ylabel('Y values', fontweight='bold')
```
