# Gross Domestic Product (GDP)

**GDP** is the total monetary market value of all the finished goods and services produced within the country borders in a specific time period. GDP acts as a comprehensive scorecard of a country's economic health, and it is a key tool for investors, and business decision makers.

GDP enables policymakers and central banks to judge whether the economy is contracting or expanding, whether it needs a boost or restraint, and if a threat such as a recession or inflation looms on the horizon.

## 1. Nominal GDP and Real GDP

GDP can be adjusted for inflation to provide deeper insights. **Real GDP** takes into account the effects of inflation while **Nominal GDP** does not. Real GDP is a better method for expressing long-term national economics performance.

**Nominal GDP** is an assessment of economic production in current prices. All goods and services counted in nominal GDP are valued at the prices that those goods and services are actually sold in that year.

**Real GDP** is an inflation-adjusted measure that reflects the number of goods and services produced by an economy in a given year, with prices held constant from year to year to separate out the impact of inflation or deflation from the trend in output over time. Since GDP is based on the monetary value of goods and services, it is subject to inflation. 

## Exercise 1: Nominal GDP and Real GDP

Let's say a country had a nominal GDP of 100 billion EUR in 2012. By 2022 its nominal GDP grew to 150 billion EUR, but prices rose 100 %. 

Let's assume that GDP growth was 5 billion EUR a year, and let's say, that the prices grew 10% each year in comparison with 2012 prices. 

**Assignment:** Make a plot showing the real and nominal change of GDP starting with 2012 and up till 2022.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

# Data Preparation

years = np.arange(2012, 2022, 1)
#print(years)
gdp = np.linspace(100, 150, 10)
#print(gdp)
#print(len(gdp))
price_index = np.linspace(1, 2, 10)
#print(price_index)
real_gdp = gdp / price_index
#print(real_gdp)

# Plot Layout
fig = plt.figure(figsize=(14, 10)) 
fig.subplots_adjust(hspace=0.35) 

gs = mpl.gridspec.GridSpec(2, 2)
ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1,:]) 

# Plot Colors
grid_color = "#d0d3d4" 
nominal_color = "#2980b9" 
real_color = "#fc712b" 


# Nominal GDP Plot
ax1.bar(years, gdp, color=nominal_color)
ax1.set_title("Nominal GDP, bln. EUR",  fontsize=18, pad=12)

# Axis Formatting
ax1.set_ylim(0, 160) # tarpas tarp stulpeliu ir grafiko virsaus, t.y. uzduodamas grafiko aukstis
ax1.set_xticks(years) # nustato kad stulpeliai bus kas metus
ax1.spines["top"].set_color(grid_color) # virsutines grafiko dezes linijos spalva
ax1.spines["right"].set_color(grid_color)
ax1.set_xlabel("Year", fontsize=12, weight="bold") # asiu antrasciu formatavimas
ax1.set_ylabel("Nominal GDP (bln. EUR)", fontsize=12, weight="bold")

# Add Horizontal Grid
ax1.grid(axis='y', color=grid_color) # groteles

# Real GDP Plot
ax2.bar(years, real_gdp, color=real_color)
ax2.set_title("Real GDP, bln. EUR (in 2012 prices)", fontsize=16, pad=10)

# Axis Formatting
ax2.set_ylim(0, 160) # tarpas tarp stulpeliu ir grafiko virsaus, t.y. uzduodamas grafiko aukstis
ax2.set_xticks(years)
ax2.spines["top"].set_color(grid_color)
ax2.spines["right"].set_color(grid_color)
ax2.set_xlabel("Year", fontsize=12, weight="bold")
ax2.set_ylabel("Real GDP (bln. EUR)", fontsize=12, weight="bold")

# Add Horizontal Grid
ax2.grid(axis='y', color=grid_color)


# Nominal GDP vs Real GDP
ax3.set_title("Nominal GDP vs Real GDP, bln. EUR", fontsize=16, pad=10)
ax3.plot(years, gdp, color=nominal_color, label="Nominal GDP")
ax3.plot(years, real_gdp, color=real_color, label="Real GDP")

# Axis Formatting
ax3.set_ylim(0, 160)
ax3.set_xticks(years)
ax3.spines["top"].set_color(grid_color)
ax3.spines["right"].set_color(grid_color)
ax3.set_xlabel("Year", fontsize=12, weight="bold")
ax3.set_ylabel("GDP (bln. EUR)", fontsize=12, weight="bold")

# Add Legend
ax3.legend()

# Add Grid
ax3.grid()

# Display Plot
plt.show()
```

## 2. GDP Per Capita



**GDP per capita** is a measurement of the GDP per person in a country’s population. It indicates that the amount of output or income per person in an economy can indicate average productivity or average living standards. GDP per capita can be nominal or real, dependant on analysis goals.

At a basic interpretation, per-capita GDP shows how much economic production value can be attributed to each individual citizen. This also translates to a measure of overall national wealth since GDP market value per person also readily serves as a prosperity measure.

## 3. GDP Growth Rate

The **GDP growth rate** compares the year-over-year change in a country’s economic output to measure how fast an economy is growing. Usually expressed as a percentage rate, this measure is popular for economic policymakers because GDP growth is thought to be closely connected to key policy targets such as inflation and unemployment rates. 

In order to calculate the GDP growth rate the Real GDP is used, this is because we are comparing the performance over time. 

## Exercise 3: Comparing GDP Growth Rates

a) Compare Estonia, Latvia, and Lithuania GDP growth rates in 2017-2022. Use grouped bar chart to complete the exercise. Use the data from the file **bc_real_gdp_growth_rates.csv**.

b) Compare Denmark, Finland, and Lithuania GDP growth rates in 2017-2022. Use grouped bar chart to complete the exercise. The growth rates data is located in **dfl_real_gdp_growth_rates.csv**.

## Solution to Exercise 3(a):

```python
data = pd.read_csv('bc_real_gdp_growth_rates.csv', index_col='Country')
#display(data.head())

years = np.array(data.columns)
countries = np.array(data.index)

country_colors = {
    
    'Estonia': '#0c6170',
    'Latvia': '#ff8370',
    'Lithuania': '#75e6da'
}

# plot
fix, ax = plt.subplots(figsize=(8, 4), layout='constrained')

x_bars = np.arange(len(years))
width = 0.3
multiplier = 0

for country in countries:
    offset = width*multiplier
    country_gdp_rate = data.loc[country] 
    cbar = ax.bar(x_bars + offset, country_gdp_rate, width, label=country, color = country_colors[country])
    ax.bar_label(cbar, padding=5)
    multiplier = multiplier +1

# plot title

ax.set_title('GDP Growth Rates in Baltic States, 2017-2022, %',fontsize=18, pad=12)

# plot legend

ax.legend(loc='upper left', ncols=len(countries))

# Y-axis

ax.set_ylim(-4, 10)
ax.set_ylabel('GDP Growth Rate, %', fontsize=12, fontweight= 'bold')

# X-axis

ax.set_xticks(x_bars + width, years)
ax.set_xlabel('Year', fontsize=12, fontweight= 'bold')

# zero axis
ax.axhline(y=0, linestyle='solid', color = 'lightgrey', linewidth=2)

plt.show()
```
